```python
from pathlib import Path
import os
import sys
import subprocess
import time
import webbrowser


# ============================================================
# PROJECT CONFIGURATION
# ============================================================

PROJECT_ROOT = Path(
    r"C:\Users\SCHOOL\Desktop\FinMark_DataPipeline"
)

PIPELINE_FILE = PROJECT_ROOT / "run_pipeline.py"
DASHBOARD_FILE = PROJECT_ROOT / "dashboard" / "app.py"

os.chdir(PROJECT_ROOT)


# ============================================================
# VERIFY REQUIRED FILES
# ============================================================

if not PIPELINE_FILE.exists():
    raise FileNotFoundError(
        f"Pipeline file not found: {PIPELINE_FILE}"
    )

if not DASHBOARD_FILE.exists():
    raise FileNotFoundError(
        f"Dashboard file not found: {DASHBOARD_FILE}"
    )


print("=" * 90)
print("FINMARK DATA ANALYTICS PIPELINE")
print("=" * 90)
print("Project Folder :", PROJECT_ROOT)
print("Python         :", sys.executable)
print("=" * 90)


# ============================================================
# RUN THE COMPLETE PIPELINE
# ============================================================

pipeline_start = time.perf_counter()

pipeline_result = subprocess.run(
    [
        sys.executable,
        "-u",
        str(PIPELINE_FILE),
    ],
    cwd=PROJECT_ROOT,
    check=False,
)

pipeline_duration = (
    time.perf_counter()
    - pipeline_start
)

print("\n" + "=" * 90)
print("PIPELINE EXECUTION RESULT")
print("=" * 90)
print(
    "Return Code      :",
    pipeline_result.returncode,
)
print(
    "Process Duration :",
    f"{pipeline_duration:.4f} seconds",
)


if pipeline_result.returncode != 0:
    print("Status           : FAILED")
    raise RuntimeError(
        "The pipeline failed. Review the stage error above."
    )

print("Status           : COMPLETED SUCCESSFULLY")
print("=" * 90)


# ============================================================
# LAUNCH THE STREAMLIT DASHBOARD
# ============================================================

dashboard_process = subprocess.Popen(
    [
        sys.executable,
        "-m",
        "streamlit",
        "run",
        str(DASHBOARD_FILE),
        "--server.headless=true",
        "--server.port=8501",
    ],
    cwd=PROJECT_ROOT,
)

time.sleep(5)

dashboard_url = "http://localhost:8501"

webbrowser.open(dashboard_url)

print("\n" + "=" * 90)
print("FINMARK DASHBOARD")
print("=" * 90)
print("Dashboard URL :", dashboard_url)
print("Process ID    :", dashboard_process.pid)
print("Status        : RUNNING")
print("=" * 90)
```
